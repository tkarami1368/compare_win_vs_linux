## writing time base on file size

```python
import matplotlib.pyplot as plt

# داده‌ها
file_sizes = windows_data['File Size (MB)']
write_time_windows = windows_data['Write Time (seconds)']
write_time_linux = linux_data['Write Time (seconds)']

# رسم نمودار
plt.figure(figsize=(10, 6))
plt.bar(file_sizes-250, write_time_windows, label='Windows Write Time',width=500 , color='blue')
plt.bar(file_sizes+250, write_time_linux, label='Linux Write Time', width=500, color='green')

plt.title('Write Time vs File Size (Windows vs Linux)')
plt.xlabel('File Size (MB)')
plt.ylabel('Write Time (seconds)')
plt.legend()
plt.grid()
plt.show()


# buffer_sizes = windows_data['Buffer Size (bytes)']

# # رسم نمودار میله‌ای
# plt.figure(figsize=(12, 6))
# plt.bar(buffer_sizes - 1000, windows_data['Read Time (seconds)'], width=2000, label='Windows Read Time', color='blue')
# plt.bar(buffer_sizes + 1000, linux_data['Read Time (seconds)'], width=2000, label='Linux Read Time', color='green')

# plt.title('Write Time vs File Size (Windows vs Linux)')
# plt.xlabel('File Size (MB)')
# plt.ylabel('Write Time (seconds)')
# plt.legend()
# plt.grid()
# plt.show()
```

```python
# Extract relevant data
file_sizes = windows_data['File Size (MB)']
buffer_sizes = windows_data['Buffer Size (bytes)']
write_times = windows_data['Write Time (seconds)']
read_times = windows_data['Read Time (seconds)']

# Unique buffer sizes for grouping
unique_buffers = sorted(buffer_sizes.unique())

# Plot bar charts for each buffer size
for buffer in unique_buffers:
    subset = windows_data[windows_data['Buffer Size (bytes)'] == buffer]

    x = subset['File Size (MB)']
    write = subset['Write Time (seconds)']
    read = subset['Read Time (seconds)']

    plt.figure(figsize=(10, 6))
    
    # Bar width
    bar_width = 0.35
    index = range(len(x))

    # Plot bars
    plt.bar(index, write, bar_width, label='Write Time (seconds)', color='pink')
    plt.bar([i + bar_width for i in index], read, bar_width, label='Read Time (seconds)', color='orange')

    # Labels and title
    plt.xlabel('File Size (MB)', fontsize=12)
    plt.ylabel('Time (seconds)', fontsize=12)
    plt.title(f'File I/O Time for Buffer Size {buffer} bytes windows', fontsize=14)
    plt.xticks([i + bar_width / 2 for i in index], x)
    plt.legend()
    plt.tight_layout()

    # Show plot
    plt.show()
```

```python
# Extract relevant data
file_sizes = linux_data['File Size (MB)']
buffer_sizes = linux_data['Buffer Size (bytes)']
write_times = linux_data['Write Time (seconds)']
read_times = linux_data['Read Time (seconds)']

# Unique buffer sizes for grouping
unique_buffers = sorted(buffer_sizes.unique())

# Plot bar charts for each buffer size
for buffer in unique_buffers:
    subset = linux_data[linux_data['Buffer Size (bytes)'] == buffer]

    x = subset['File Size (MB)']
    write = subset['Write Time (seconds)']
    read = subset['Read Time (seconds)']

    plt.figure(figsize=(10, 6))
    
    # Bar width
    bar_width = 0.35
    index = range(len(x))

    # Plot bars
    plt.bar(index, write, bar_width, label='Write Time (seconds)', color='red')
    plt.bar([i + bar_width for i in index], read, bar_width, label='Read Time (seconds)', color='black')

    # Labels and title
    plt.xlabel('File Size (MB)', fontsize=12)
    plt.ylabel('Time (seconds)', fontsize=12)
    plt.title(f'File I/O Time for Buffer Size {buffer} bytes in linux', fontsize=14)
    plt.xticks([i + bar_width / 2 for i in index], x)
    plt.legend()
    plt.tight_layout()

    # Show plot
    plt.show()
```
